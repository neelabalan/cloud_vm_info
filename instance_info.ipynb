```python
import urllib.parse
import requests
import pandas as pd

```

```python
region_name = 'US East (Ohio)'
aws_ec2_url = f'https://b0.p.awsstatic.com/pricing/2.0/meteredUnitMaps/ec2/USD/current/ec2-ondemand-without-sec-sel/{urllib.parse.quote_plus(region_name)}/Linux/index.json'
```

```python
response = requests.get(aws_ec2_url)
response
```

```python
response_json = response.json()
response_json
```

```python
response_json.keys()
```

```python
len(response_json['regions'][region_name])
```

```python
aws_info_df = pd.DataFrame(response_json['regions'][region_name].values())
```

```python
aws_info_df.head()
```

```python
aws_info_df.columns
```

```python
aws_info_df.dtypes
```

```python
aws_info_df['Memory'] = aws_info_df['Memory'].str.replace('GiB', '')
```

```python
aws_info_df['vCPU'] = aws_info_df['vCPU'].astype(int)
aws_info_df['Memory'] = aws_info_df['Memory'].astype(float)
aws_info_df['price'] = aws_info_df['price'].astype(float)
```

```python
aws_info_df[['vCPU', 'Memory', 'price']].describe()
```

```python
aws_info_df.to_csv('data/aws_ec2.csv')
```
